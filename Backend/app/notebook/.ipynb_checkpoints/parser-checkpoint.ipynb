```python
import fitz  # PyMuPDF
import json

def simple_parse_pdf(pdf_path = "../data/data.pdf"):
    doc = fitz.open(pdf_path)

    chunks = []

    for page_num, page in enumerate(doc):
        text = page.get_text("text")
        text = text.strip()

        if text:
            chunks.append({
                "page": page_num + 1,
                "content": text
            })

    return chunks
```

```python
document = simple_parse_pdf()
```

```python
nlp = English()

nlp.add_pipe("sentencizer")
for item in tqdm(document, desc="Processing Pages"):
    # Split content into sentences using spaCy
    doc = nlp(" ".join(item['content']))
    item['sentences'] = [str(sent) for sent in doc.sents]
    
    # Store the number of sentences
    item['page_sentence_count_spacy'] = len(item['sentences'])
```

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin, urlparse

visited = set()

def crawl(url, base_url):
    if url in visited:
        return
    visited.add(url)
    try:
        response = requests.get(url)
        if response.status_code == 200:
            soup = BeautifulSoup(response.text, "html.parser")
            for link_tag in soup.find_all('a', href=True):
                href = link_tag['href']
                full_url = urljoin(base_url, href)
                if urlparse(full_url).netloc == urlparse(base_url).netloc:
                    crawl(full_url, base_url)

    except Exception as e:
        print(f"Error visiting {url}: {e}")

def start_crawling(base_url):
    crawl(base_url, base_url)
    return list(visited)
```

```python
base_url = 'https://www.insightglobal.com'  # Replace with your desired base URL
urls = start_crawling(base_url)
```

```python
import requests
from bs4 import BeautifulSoup

# Define fetch and parse function
def fetch_and_parse(url):
    try:
        headers = {
            "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"  # Pretend to be a real browser
        }
        response = requests.get(url, headers=headers, timeout=10)
        response.raise_for_status()

        soup = BeautifulSoup(response.text, 'html.parser')

        # Remove scripts and styles
        for script_or_style in soup(["script", "style"]):
            script_or_style.decompose()

        # Get clean text
        text = soup.get_text(separator=" ", strip=True)
        return text

    except Exception as e:
        print(f"Error fetching {url}: {e}")
        return ""


# Now fetch contents properly
contents = []
for url in urls:
    content = fetch_and_parse(url)
    if content:  # Only append if content is not empty
        contents.append({
            "url": url,
            "content": content
        })

# Now print nicely
for item in contents:
    print(f"URL: {item['url']}")
    print(f"Content Preview: {item['content'][:300]}...")  
    print("-" * 80)
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import time

def fetch_and_parse_with_selenium(url):
    try:
        # Setup Chrome options
        chrome_options = Options()
        chrome_options.add_argument("--headless")  # Run headless
        chrome_options.add_argument("--no-sandbox")
        chrome_options.add_argument("--disable-dev-shm-usage")

        # Setup service
        service = Service(ChromeDriverManager().install())

        # Setup the driver properly
        driver = webdriver.Chrome(service=service, options=chrome_options)

        # Open the URL
        driver.get(url)
        time.sleep(3)  # wait for JavaScript to load

        # Get page source
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')

        # Clean the soup
        for script_or_style in soup(["script", "style"]):
            script_or_style.decompose()

        text = soup.get_text(separator=" ", strip=True)

        driver.quit()

        return text

    except Exception as e:
        print(f"Error fetching {url}: {e}")
        return ""
```

```python
contents = []
for url in urls:
    content = fetch_and_parse_with_selenium(url)
    if content:
        contents.append({
            "url": url,
            "content": content
        })

# Now print or save
for item in contents:
    print(f"URL: {item['url']}")
    print(f"Content Preview: {item['content'][:300]}...")
    print("-" * 80)
```
